After importing, merge the 3 data sets into 1 single data set called **nat**. *You must use the options statement to prevent extaneous SAS output for this problem.* (SAS, Accessing and Importing Data, Data Wrangling)

  - M_Ht_In

```sas
filename nat1 url "https://uofi.box.com/shared/static/gb07nz7jafspk65wzxq6awlu02kypiq7.csv";
filename nat10 url "https://uofi.box.com/shared/static/xf7r94alpzaqv4wxh5liqw7rpk21wb6l.csv";
filename nat20 url "https://uofi.box.com/shared/static/eq9k4d73ansal5p8mspu3mpslmzs7x8g.csv";

data natpart1;
    infile nat1 dsd firstobs=2 missover;
    input M_Ht_In BMI PWgt_R DWgt_R RF_PDIAB $ RF_GDIAB $ RF_PHYPE $ RF_GHYPE $ 
          RF_EHYPE $ RF_PPTERM $ RF_CESAR $ ME_PRES ATTEND PAY_REC APGAR5 APGAR10 AB_NICU $
          BFED $ priorlive priordead mrace6 frace6 dplural$ OEGest_R3 SEX $ MAGER FAGECOMB MEDUC 
          FEDUC PREVIS_REC previs CIG_REC$ WTGAIN DMETH_REC dbwt 
run;

data natpart10;
    infile nat10 dsd firstobs=2 missover;
    input M_Ht_In BMI PWgt_R DWgt_R RF_PDIAB $ RF_GDIAB $ RF_PHYPE $ RF_GHYPE $ 
          RF_EHYPE $ RF_PPTERM $ RF_CESAR $ ME_PRES ATTEND PAY_REC APGAR5 APGAR10 AB_NICU $
          BFED $ priorlive priordead mrace6 frace6 dplural$ OEGest_R3 SEX $ MAGER FAGECOMB MEDUC 
          FEDUC PREVIS_REC previs CIG_REC$ WTGAIN DMETH_REC dbwt 
run;

data natpart20;
    infile nat20 dsd firstobs=2 missover;
    input M_Ht_In BMI PWgt_R DWgt_R RF_PDIAB $ RF_GDIAB $ RF_PHYPE $ RF_GHYPE $ 
          RF_EHYPE $ RF_PPTERM $ RF_CESAR $ ME_PRES ATTEND PAY_REC APGAR5 APGAR10 AB_NICU $
          BFED $ priorlive priordead mrace6 frace6 dplural$ OEGest_R3 SEX $ MAGER FAGECOMB MEDUC 
          FEDUC PREVIS_REC previs CIG_REC$ WTGAIN DMETH_REC dbwt 
run;

data nat;
    set natpart1 natpart10 natpart20;
run;
proc contents data=nat;
run;
```
